```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By

# 교보문고 컴퓨터/IT 스테디셀러 50권
# 2024.11.11 ~ 2024.11.17
# 책 제목, 저자, 가격, 출판일, 별점, 리뷰수, 할인율
driver = webdriver.Chrome()

url = 'https://store.kyobobook.co.kr/bestseller/steady/domestic/33?page=1&per=50'
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
import re

it_books = []

book_items_10 = soup.select('body > div.relative.min-w-\[1440px\] > main > section > div > div > section > ol.grid.border-t.border-gray-400.grid-cols-1.pt-9 > li')

for book_item in book_items_10:
    title = book_item.select_one('div.ml-4.min-w-\[516px\].w-full > a').text.strip()
    info_div = book_item.select_one('div.line-clamp-2.flex.overflow-hidden.whitespace-normal.break-all.text-gray-800.font-text-l.mt-1')
    if info_div:
        info_text = info_div.get_text(" ", strip=True)  
        info_parts = info_text.split(' · ')  # ' · ' 기준으로 분리
        author = info_parts[0].strip()
        publisher = info_parts[1].split('·')[0].strip()

    price_element = book_item.select_one('div.flex.flex-col.mt-3 > div > span:nth-child(2)')
    price = price_element.text.strip()
    price = re.sub(r'[^\d]', '', price)
    publish_date = book_item.select_one('div.line-clamp-2.flex.overflow-hidden.whitespace-normal.break-all.text-gray-800.font-text-l.mt-1 > span').text.strip() 
    publish_date = publish_date.replace('·\xa0', '').strip()
    rating = book_item.select_one('div.flex.w-full.flex-wrap.gap-3 > div > div > div > div > span.font-weight-bold.text-black').text.strip() 
    reviews = book_item.select_one('div.flex.w-full.flex-wrap.gap-3 > div > div > div > div > span.font-weight-regular.text-gray-700').text.strip() 
    reviews = re.findall(r'\d+', reviews)
    reviews = reviews[0]
    discount = book_item.select_one('div.flex.flex-col.mt-3 > div > span.font-weight-bold.inline-block.align-top.text-green-800.font-text-xl').text.strip() 

    it_books.append({
        'title': title,
        'author': author,
        'publisher' : publisher,
        'price': price,
        'publish_date': publish_date,
        'rating': rating,
        'reviews': reviews,
        'discount': discount
    })

for book in it_books:
    print(book)
```

```python
book_items_40 = soup.select('body > div.relative.min-w-\[1440px\] > main > section > div > div > section > ol.grid.grid-cols-1.pt-0 > li')

for book_item in book_items_40:
    title = book_item.select_one('body > div.relative.min-w-\[1440px\] > main > section > div > div > section > ol.grid.grid-cols-1.pt-0 > li > div > div.flex.items-top.justify-start.mr-\[35px\].w-full > div.ml-4.min-w-\[516px\].w-full > a').text.strip()
    info_div = book_item.select_one('div.line-clamp-2.flex.overflow-hidden.whitespace-normal.break-all.text-gray-800.font-text-l.mt-1')
    if info_div:
        info_text = info_div.get_text(" ", strip=True)  
        info_parts = info_text.split(' · ')  # ' · ' 기준으로 분리
        author = info_parts[0].strip()
        publisher = info_parts[1].split('·')[0].strip()
    
    price_element = book_item.select_one('div.flex.flex-col.mt-3 > div > span:nth-child(2) > span.font-weight-bold')
    if not price_element:
        price_element = book_item.select_one('body > div.relative.min-w-\[1440px\] > main > section > div > div > section > ol.grid.grid-cols-1.pt-0 > li:nth-child(40) > div > div.flex.items-top.justify-start.mr-\[35px\].w-full > div.ml-4.min-w-\[516px\].w-full > div.flex.flex-col.mt-3 > div > span.inline-block.align-top.font-text-xl > span.font-weight-bold')
    
    if price_element:
        price = price_element.text.strip()
        price = re.sub(r'[^\d]', '', price)
    else:
        price = '가격 없음'

    publish_date = book_item.select_one('div.line-clamp-2.flex.overflow-hidden.whitespace-normal.break-all.text-gray-800.font-text-l.mt-1 > span').text.strip() 
    publish_date = publish_date.replace('·\xa0', '').strip()
    rating = book_item.select_one('div.flex.w-full.flex-wrap.gap-3 > div > div > div > div > span.font-weight-bold.text-black').text.strip() 
    reviews = book_item.select_one('div.flex.w-full.flex-wrap.gap-3 > div > div > div > div > span.font-weight-regular.text-gray-700').text.strip() 
    reviews = re.findall(r'\d+', reviews)
    reviews = reviews[0]
    discount_element = book_item.select_one('li > div > div.flex.items-top.justify-start.mr-\[35px\].w-full > div.ml-4.min-w-\[516px\].w-full > div.flex.flex-col.mt-3 > div > span.font-weight-bold.inline-block.align-top.text-green-800.font-text-xl')
    discount = discount_element.text.strip() if discount_element else '할인 없음'
    it_books.append({
        'title': title,
        'author': author,
        'publisher' : publisher,
        'price': price,
        'publish_date': publish_date,
        'rating': rating,
        'reviews': reviews,
        'discount': discount
    })

for book in it_books:
    print(book)
```

```python
import pandas as pd

df = pd.DataFrame(it_books)
df.head()
```

```python
df.to_csv('it_books.csv', header=True, index=False)
```

```python
import pymysql

connection = pymysql.connect(
    host='localhost',
    user='root',  
    password='0000',  
    database='my_db'  
)

cursor = connection.cursor()

# 기존 테이블 삭제
drop_table_query = "DROP TABLE IF EXISTS books_table"
cursor.execute(drop_table_query)

create_table_query = """
CREATE TABLE IF NOT EXISTS books_table (
    title VARCHAR(255),
    author VARCHAR(255),
    publisher VARCHAR(255),
    price VARCHAR(255),
    publish_date VARCHAR(255),
    rating VARCHAR(255),
    reviews INT,
    discount VARCHAR(255)
)
"""
cursor.execute(create_table_query)

for index, row in df.iterrows():
    insert_query = """
    INSERT INTO books_table (title, author, publisher, price, publish_date, rating, reviews, discount)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """
    cursor.execute(insert_query, (
        row['title'], row['author'], row['publisher'], row['price'], 
        row['publish_date'], row['rating'], row['reviews'], row['discount']
    ))

connection.commit()
cursor.close()
connection.close()
```
